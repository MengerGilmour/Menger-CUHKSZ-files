# **Q2: Palindromic primes**
tu4_q2.PNG

```
from math import sqrt
def isPrime(num):
  if num <= 1:
    return
  # Problem: how to judge if a number is prime
  for i in range(2, int(sqrt(num))+1):
    if num % i == 0:
      return False

  return True

def reverse(number):
  num_str = str(number)
  # num_str is list of characters
  num_str_reverse = num_str[::-1]
  num_reverse = int(num_str_reverse)
  return num_reverse

def isPalindrome(number):
  num_reverse = reverse(number=number)
  if num_reverse == number:
    return True
  else:
    return False
```

```
from time import time
prime_palin = []
target_number = 100

prime_palin.append(2)
number = 3
t1 = time()
while len(prime_palin) < target_number:
  # and operator: calc the 1st expression
  # if it is False, it will NOT calc the 2rd expression
  # if isPalindrome(number) and isPrime(number):
  if isPrime(number) and isPalindrome(number):
    prime_palin.append(number)
  else:
    # do nothing
    pass

  # we only need to check odd number
  number += 2

print('Finished, time cost: ', time()-t1)
# formatted print
def special_print(numbers, num_per_row):
  # numbers is a list
  # num_per_row: number of elements in a for each row
  for k, i in enumerate(numbers):
    print('%8d' % i, end='')
    # print('{:.8d}'.format(i), end='')
    if (k+1) % num_per_row == 0:
      # it means just print '\n'
      print()

special_print(prime_palin, num_per_row=10)
```

# **Q3: Mytriangle module**
tu4_q3.PNG

```
from math import sqrt
def isValid(a, b, c):
  if a>0.0 and b>0.0 and c>0.0:
    if a+b>c and a+c>b and b+c>a:
      return True

  return False

def area(a, b, c):
  if isValid(a, b, c):
    s = (a+b+c)/2.0
    return sqrt(s*(s-a)*(s-b)*(s-c))


inputs = input('Please input three floats, separated by ,: ')
inputs = inputs.split(',')

s1 = float(inputs[0])
s2 = float(inputs[1])
s3 = float(inputs[2])
# Better way
# s1, s2, s3 = map(float, inputs)
print('Area=', area(s1,s2,s3))
```

# **Q4: Yang Hui’s Triangle**
tu4_q4.PNG

```
def factorial(n):
  if n == 0:
    return 1

  if n < 0:
    print('??')
    return

  res = 1
  for i in range(1,n+1):
    # like res = res*i, but a inplace operation
    res *= i

  return res

def combinatorial_number(n, k):
  if k < 0 or k > n or n < 0:
    print('??')
    return

  if n == 0:
    return 1

  return int(factorial(n) / (factorial(k)*factorial(n-k)))

cn = combinatorial_number
row = int(input('Please input the row number: '))
for i in range(row+1):
  # number of rows: row+1
  # so it is a [row+1, 2*row+1] matrix
  for j in range(-row, row+1):
    k = (i - abs(j))
    if abs(j) <= i and k % 2 == 0:
      # in the triangle
      k = k // 2
      print('%6d' % cn(i, k), end='')
    else:
      print('%6s' % '', end='')

  # after print a row, print \n
  print()
```
